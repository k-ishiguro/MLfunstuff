```python
import numpy as np
import scipy.special as ss
import matplotlib.pylab as plt
```

```python
def beta(a, b, mew):
    #three gamma funcs for normalizer
    e1 = ss.gamma(a + b)
    e2 = ss.gamma(a)
    e3 = ss.gamma(b)
    # responcses
    e4 = mew ** (a - 1)
    e5 = (1 - mew) ** (b - 1)
    return (e1/(e2*e3)) * e4 * e5
```

```python
def plot_beta(a, b, figid):
    Ly = []
    Lx = []
    # take dense grids
    mews = np.mgrid[0.00001:1-0.00001:100j]
    # inefficient but easy
    for mew in mews:
        Lx.append(mew)
        Ly.append(beta(a, b, mew))
    fig = plt.figure(figid)
    plt.plot(Lx, Ly, label="a=%f, b=%f" %(a,b))
    plt.xlim(0.0, 1.0)
    plt.ylim(0.0, 3.0)
    plt.legend()
      
    outname = "beta_a" + str(a) + "_b" + str(b)
    plt.title(outname + ', PDF')
    
    plt.savefig(outname + '.eps')
    plt.savefig(outname + '.png')
```

```python
plot_beta(0.5, 0.5, 1)
plot_beta(2.0, 2.0, 2)
plot_beta(1.0, 1.0, 3)
```
